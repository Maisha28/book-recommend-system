```python
import numpy as np
import pandas as pd
```

```python
books=pd.read_csv('Books.csv')
users=pd.read_csv('Users.csv')
ratings=pd.read_csv('Ratings.csv')
```

```python
#popularity based recommendation system
#basically atleast x no of ratings and highest ratings
```

```python
ratings_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df= ratings_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_name.groupby(
    'Book-Title',
    as_index=False
)['Book-Rating'].mean()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
```

```python
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df=popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False)
```

```python
popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```
